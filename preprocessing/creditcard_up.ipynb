```python
import pandas as pd
import numpy as np
from sklearn.utils import resample
```

```python
dataset = pd.read_csv('required_datasets/creditcard.csv')
dataset_up=pd.read_csv('required_datasets/upsampling sorted_time.csv')
dataset_up.head()
```

```python
negative_class = dataset_up[dataset_up['Class'] == 1]

positive_class = dataset[dataset['Class'] == 0]
```

```python
# Downsample the positive class
downsampled_positive = resample(positive_class,
                                replace=False,  # sample without replacement
                                n_samples=len(negative_class),  # match the number of samples in the negative class
                                random_state=0) 

# Combine the downsampled positive class with the negative class
balanced_data = pd.concat([downsampled_positive, negative_class])

# Shuffle the data to ensure the order is randomized
balanced_data = balanced_data.sample(frac=1, random_state=42)

# Save the balanced dataset to a new CSV file
# balanced_data.to_csv('upsample_creditcard.csv', index=False)
```

```python
dataset=balanced_data
```

```python
# dataset=pd.read_csv("upsample_creditcard.csv")

from sklearn.preprocessing import StandardScaler

dataset4 = dataset
# Normalizace the Amount between -1,1
dataset4['normalizedAmount'] = StandardScaler().fit_transform(dataset4['Amount'].values.reshape(-1,1))

a=dataset4['Class']
b=dataset4['normalizedAmount']

dataset4 = dataset4.drop(columns = ['Class','normalizedAmount'])

dataset4['normalizedAmount']=b
dataset4['Class']=a


dataset4 = dataset4.drop(columns = ['Amount','Time'])
dataset4.head()
```

```python
X = dataset4.drop(columns = ['Class'])
y = dataset4['Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
