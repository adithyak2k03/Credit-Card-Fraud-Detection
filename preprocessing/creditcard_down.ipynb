```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("required_datasets/creditcard.csv")
df.shape
```

```python
positive_class = df[df['Class'] == 0]
negative_class = df[df['Class'] == 1]
```

```python
from sklearn.utils import resample
```

```python
# Downsample the positive class
downsampled_positive = resample(positive_class,
                                replace=False,  # sample without replacement
                                n_samples=len(negative_class)+100,  # match the number of samples in the negative class
                                random_state=42)  # set random seed for reproducibility

# Combine the downsampled positive class with the negative class
balanced_data = pd.concat([downsampled_positive, negative_class])

# Shuffle the data to ensure the order is randomized
balanced_data = balanced_data.sample(frac=1, random_state=42)

# Save the balanced dataset to a new CSV file


# balanced_data.to_csv('your_balanced_dataset.csv', index=False)
```

```python
dataset=balanced_data
```

```python
from sklearn.preprocessing import StandardScaler

dataset3 = dataset
# Normalizace the Amount between -1,1
dataset3['normalizedAmount'] = StandardScaler().fit_transform(dataset3['Amount'].values.reshape(-1,1))

a=dataset3['Class']
b=dataset3['normalizedAmount']

dataset3 = dataset3.drop(columns = ['Class','normalizedAmount'])

dataset3['normalizedAmount']=b
dataset3['Class']=a


dataset3 = dataset3.drop(columns = ['Amount','Time'])
dataset3.head()
```

```python
X = dataset3.drop(columns = ['Class'])
y = dataset3['Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
